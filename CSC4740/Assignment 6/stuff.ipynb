```python
def distance(p1, p2):
    return math.sqrt(math.pow((p2[0]-p1[0]), 2) + math.pow(p2[1]-p1[1], 2))
```

```python
def closer_to(p1):
    c1 = (1.8, 5.5)
    c2 = (7, 4.3)
    if distance(p1, c1) <= distance(p1, c2):
        return(f"{p1} is closest to {c1}")
    else:
        return(f"{p1} is closest to {c2}")
```

```python
def dist_to(p1):
    c1 = (1.8, 5.5)
    c2 = (7, 4.3)
    return(f"Distance of {p1} to {c1}: {distance(p1, c1)}, {c2}: {distance(p1, c2)}")
```
