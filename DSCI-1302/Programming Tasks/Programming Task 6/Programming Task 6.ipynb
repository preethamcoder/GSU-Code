Q1. Use matplotlib to draw a same plot as Q1.jpg (attached). 
Requirements:
1. (20%) plot two curves: **2sin(x)** in solid line and **2cos(x)** in dotted line
2. (10%) plot three periods for each function as shown in Q1.jpg
3. (10%) plot legend, axis labels (font size is 15), and grid (color is blue, alpha=0.2)
4. (10%) the range of y-axis is [-4, 4]
5. (10%) color: use red when the function is increasing and use green when it is decreasing.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def f(x):
    return 2*np.sin(x)
#Although I could define a separate cos(x) function, I used the trig identity sin(pi/2 - x) = cos(x)
dom = np.linspace(-2.5*math.pi, 3.5*math.pi, 200)
dom1 = np.linspace(-3*math.pi, 3*math.pi, 200)
vals = [f(i) for i in dom]
vals2 = [f(math.pi-i) for i in dom]
plt.plot(dom[:34], vals[:34], "r", dom1[:34], vals2[:34], "r--", dom1[34:68], vals2[34:68], "g--", dom1[68:102], vals2[68:102], "r--", dom1[100:134], vals2[100:134], "g--", dom1[134:166], vals2[134:166], "r--", dom1[166:], vals2[166:], "g--", dom[34:68], vals[34:68], "g",
         dom[68:102], vals[68:102], "r", dom[100:134], vals[100:134], "g", dom[134:166], vals[134:166], "r", dom[166:], vals[166:], "g")
plt.ylim(-4, 4)
plt.xlabel("X", fontsize=16)
plt.ylabel("Y", fontsize=16)
plt.legend(['2sin(x)','2cos(x)'])
plt.grid(color="lavender", linestyle="-", linewidth=1.15)
plt.show()
```
