Q1. Read in the "unicode_count.txt" file (attached), and save it as a dictionary. You will use this data for all the questions below. (10 pts)

Plot the top 10 most frequent words in a bar chart. (10 pts)

Add necessary labels, legends, titles, etc to make it look nicer. (10 pts)

**Hint**. "unicode_count.txt" is generated from Programming Task 3 Q3. It lists all the words in an article with their occurrences. 

```python
import matplotlib.pyplot as plt
f = open("unicode_count.txt", "r")
lines = f.readlines()
ind_lines = []
for i in lines:
    ind_lines.append(i.split())
words_occ = {}
for i in ind_lines:
    words_occ[i[0]] = int(i[1])
sorted_occ_keys = sorted(words_occ, key = words_occ.get, reverse=True)
final_dict = {}
for i in sorted_occ_keys:
    final_dict[i] = words_occ[i]
x_ax = []
y_ax = []
for i in final_dict:
    x_ax.append(i)
    y_ax.append(final_dict[i])
plt.bar(x_ax[:10], y_ax[:10], width=0.9, color="green")
plt.grid(color = "brown")
plt.title("Top 10 words in unicode_count.txt")
plt.xlabel("Word")
plt.ylabel("Occurence")
plt.text(5, 25, "1 square = 5 occurences", style="italic", bbox={'facecolor':'yellow'})
plt.show()
```

Q2. Plot a scatter chart. The x-axis denotes the initial letter of the word, and the y-axis denotes the length of the word. (10 pts)

Use different colors to denote the frequency. Use the "summer" color map. Add a color bar to the right of the plot. (10 pts)

Add necessary labels, legends, titles, etc to make it look nicer. (10 pts)

```python
lett = []
lengths = []
for i in x_ax:
    lett.append(i[0])
    lengths.append(len(i))
plt.scatter(lett[183:], lengths[183:], c="greenyellow")
plt.scatter(lett[121:182], lengths[121:182], c="springgreen")
plt.scatter(lett[60:121], lengths[60:121], c="forestgreen")
plt.scatter(lett[:60], lengths[:60], c="limegreen")
plt.summer()
plt.colorbar()
plt.title("Scatterplot of starting letters and lengths of words")
plt.ylabel("Word Length")
plt.xlabel("Starting Letters")
plt.show()
```
