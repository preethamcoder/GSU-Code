```python
import matplotlib.pyplot as plt
scores = [1, 3, 2, 23, 8, 24, 49, 13, 75, 0, 31, 6, 14]
years = list(range(2008, 2021))
plt.plot(years, scores, color = "green")
plt.bar(years, scores, color = "red")
plt.ylabel("Runs")
plt.xlabel("Years")
plt.title("Virat Kohli's performances in RCB's first game over the years")
plt.show()
```

```python
scores2 = [0, 7, 54, 64, 0, 0, 28, 82, 0, 44, 9, 51]
years2 = list(range(2009, 2021))
plt.plot(years2, scores2, color = "green")
plt.bar(years2, scores2, color = "yellow")
plt.ylabel("Runs")
plt.xlabel("Years")
plt.title("ABD's performances in first games over the years (didn't bat in '09, 13, '14, and '17)")
plt.show()
```
