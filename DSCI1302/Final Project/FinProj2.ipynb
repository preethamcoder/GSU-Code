```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
data = pd.read_csv("csv_Files/forestfires.csv")
data.describe()
#data
```

```python
data.notnull().sum()
```

```python
ffmc = np.array(data['FFMC'].tolist())
dmc = np.array(data['DMC'].tolist())
isi = np.array(data['ISI'].tolist())
temp = np.array(data['temp'].tolist())
area = np.array(data['area'].tolist())
wind = np.array(data['wind'].tolist())
relhum = np.array(data['RH'].tolist())
rain = np.array(data['rain'].tolist())
dc = np.array(data['DC'].tolist())
X = range(1, 518)
plt.bar(ffmc, area)
plt.grid(color="pink")
plt.title("FFMC and Area co-relation")
plt.xlabel("FFMC")
plt.ylabel("Area burned in hectares")
#plt.bar(X, area)
#area.mean(), temp.mean(), ffmc.mean()
#data
plt.show()
```

```python
ind = area >= 19.99
plt.plot(relhum[~ind], area[~ind], "yo")
plt.plot(relhum[ind], area[ind], "ro")
#plt.bar(X, data['RH'].tolist())
plt.ylabel("Area burned in hectares")
plt.xlabel("Relative Humidity in % - 15.00-100.00")
plt.title("Humidity and Area burned")
plt.show()
```

```python
ind = dc >=450
plt.scatter(dc[~ind], area[~ind], color="yellow")
plt.scatter(dc[ind], area[ind], color="maroon")
plt.xlabel("Drought Content Index")
plt.ylabel("Area burned in hectares")
plt.title("DC's impact on Area burned")
plt.show()
```

```python
#rain2 = [i for i in rain if(i > 0)]
#ind = [rain.index(i) for i in rain if(i > 0)]
#area2 = [area[i] for i in ind]
#rain2, area2
ind2 = area > 10.0
#plt.plot(rain[ind2], area[ind2], "ro")
#plt.show()
plt.plot(rain[~ind], area[~ind], "yo")
plt.plot(rain[ind], area[ind], "ro")
plt.xlabel("Rainfall (mm/$m^2$)")
plt.ylabel("Area burned in hectares")
plt.title("Rainfall's impact on area burned")
plt.show()
```

```python
#plt.plot(dmc, area, "go")
plt.bar(temp, area)
plt.xlabel("Temperature in Celcius")
plt.ylabel("Area burned in hectares")
plt.title("Temperature vs Area Burned")
plt.grid(color="green")
plt.show()
```

```python
ind = wind >= 3.25
plt.scatter(wind[~ind], area[~ind], color="red")
plt.scatter(wind[ind], area[ind], color="yellow")
plt.xlabel("Speed of wind in kmph")
plt.ylabel("Area burned in hectares")
plt.title("Wind and it's impact on area burned")
plt.show()
```
