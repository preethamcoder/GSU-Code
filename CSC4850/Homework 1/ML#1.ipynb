```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("image.png")
U, s, vh = np.linalg.svd(img)
```

```python
img
```

```python
U.shape, s.shape, vh.shape
```

```python
def reconstruction(U, s, vh, k):
    return np.dot(np.dot(U[:, :k], np.diag(s[:k])),vh[:k])
```

```python
values = [1, 5, 10, 20, 40]
for each in values:
    plt.imshow(reconstruction(U, s, vh, each))
    plt.savefig(f"image with {each} values.jpg")
```
